```python
import pandas as pd
from sklearn.datasets import load_boston
```

```python
dt =load_boston()

dataset=pd.DataFrame(dt.data,columns=load_boston().feature_names)
```

```python
#pergunta 1

dataset.head()
```

```python
#pergunta 2

dataset.shape
```

```python
#pergunta 3

feat = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B']

#pergunta 6

x =dataset[feat]
x.head()
```

```python
#pergunta 4

#pergunta 6

y=dataset['LSTAT']
y.head()
```

```python
#pergunta 5

import seaborn as sns
%matplotlib inline

sns.pairplot(dataset, x_vars=feat, y_vars='LSTAT', size=7, aspect=0.7, kind='reg')
```

```python
#pergunta 7

print(type(x))
print(x.shape)

print(type(y))
print(y.shape)
```

```python
# pergunta 9

from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# pergunta 8
#base de dados sobre preços de casas
```

```python
# pergunta 10
# 12%
```

```python
# pergunta 11

from sklearn.linear_model import LinearRegression

linreg = LinearRegression()
linreg.fit(x_train, y_train)
```

```python
# pergunta 12

print(linreg.intercept_)
print(linreg.coef_)
```

```python
# pergunta 13

list(zip(feat, linreg.coef_))
```

```python
# pergunta 14

y_prev = linreg.predict(x_test)
```

```python
#pergunta 15

from sklearn import metrics
import numpy as np

print(np.sqrt(metrics.mean_squared_error(y_test, y_prev)))
```

```python
# pergunta 16

feat0 = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B']
feat1 = ['CRIM', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'RAD', 'PTRATIO', 'B']

x = dataset[feat1]
y=dataset['LSTAT']

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)

linreg.fit(x_train, y_train)

y_prev = linreg.predict(x_test)
print(np.sqrt(metrics.mean_squared_error(y_test, y_prev)))
```
